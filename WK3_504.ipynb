## SQL Table Creation

### Creating from Scratch

```python
# Initiate Connection

conn = sqlite3.connect('health.db')
c = conn.cursor()
```

```python
# Create Table

c.execute('''
    CREATE TABLE price_chart
      (
        [Code] real,
        [Type] text,
        [Package/Line_Level] text,
        [Gross_charge] real,
        [Discounted_cash_price] real

      );

''')

conn.commit()
```

```python
# Insert Into Manual

sql_query1 = """
    INSERT INTO price_chart (
      'Code',
      'Type',
      'Package/Line_Level',
      'Gross_charge',
      'Discounted_cash_price'
    )
    values (
      42416,
      'Inpatient',
      'Package',
      500.00,
      400.00
    );
"""

c.execute(sql_query1)

```

```python
# Second Insert

sql_query2 = """
    INSERT INTO price_chart (
      'Code',
      'Type',
      'Package/Line_Level',
      'Gross_charge',
      'Discounted_cash_price'
    )
    values (
      85781,
      'Outpatient',
      'Line',
      1000.00,
      750.00
    );
"""

c.execute(sql_query2)
```

```python
# Test to see if inserted

sql_query3 = '''
SELECT *
FROM price_chart;
'''

c.execute(sql_query3)
print(c.fetchall())
```

### Automatic Insertion from Dataset

```python
# Inserting smaller portion of dataset

df1_small = df1.head(1000)

df1_small.to_sql('price_chart', conn, if_exists='replace')
```
